```python
import random 
import numpy as np 
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
from sklearn.datasets.samples_generator import make_blobs 
%matplotlib inline
from sklearn import datasets
from sklearn.decomposition import PCA
from scipy import ndimage 
from scipy.cluster import hierarchy 
from scipy.spatial import distance_matrix 
from sklearn.cluster import DBSCAN
from sklearn.metrics import accuracy_score
```

```python
iris = datasets.load_iris()
```

```python
iris
```

```python
X = iris.data  
y = iris.target

```

```python
X
```

```python
X_reduced = PCA(n_components=3).fit_transform(iris.data)
X_reduced
```

```python
np.random.seed(0)
clustering = DBSCAN(eps=0.6, min_samples=3).fit(X_reduced)
labels = clustering.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
n_clusters_
```

```python
print(labels == y)
```

```python
accuracy_score(labels, y)
```

```python
np.random.seed(0)
k_means=KMeans(n_clusters=3, init='k-means++', n_init=1)
k_means=k_means.fit(X_reduced)
k_means_labels=k_means.labels_
k_means_labels
```

```python
accuracy_score(k_means_labels, y)
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python

agglom = AgglomerativeClustering(n_clusters=3, linkage='complete')
agglom = agglom.fit(X_reduced)
label = agglom.labels_
label
```

```python
accuracy_score(label, y)
```

```python
X_reduced1 = PCA(n_components=2).fit_transform(iris.data)
np.random.seed(0)
clustering = DBSCAN(eps=0.6, min_samples=3).fit(X_reduced1)
labels = clustering.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
print(n_clusters_)
print("dbscan:")
print(accuracy_score(labels, y))
k_means=KMeans(n_clusters=3, init='k-means++', n_init=1)
k_means=k_means.fit(X_reduced1)
k_means_labels=k_means.labels_
print("k means:")
print(accuracy_score(k_means_labels, y))
agglom = AgglomerativeClustering(n_clusters=3, linkage='complete')
agglom = agglom.fit(X_reduced1)
label = agglom.labels_
print("heia")
print(accuracy_score(label, y))



```

```python
a=PCA(n_components=1)
X_reduced2 = a.fit_transform(iris.data)
np.random.seed(0)
clustering = DBSCAN(eps=0.6, min_samples=3).fit(X_reduced2)
labels = clustering.labels_
n_clusters_ = len(set(labels)) - (1 if -1 in labels else 0)
print(n_clusters_)
print("dbscan:")
print(accuracy_score(labels, y))
k_means=KMeans(n_clusters=3, init='k-means++', n_init=1)
k_means=k_means.fit(X_reduced2)
k_means_labels=k_means.labels_
print("k means:")
print(accuracy_score(k_means_labels, y))
agglom = AgglomerativeClustering(n_clusters=3, linkage='complete')
agglom = agglom.fit(X_reduced2)
label = agglom.labels_
for i in range(len(label)):
    if label[i] == 0 :
        label[i] = 1
    elif label[i] == 1:
        label[i] = 0
print(label)
print(y)
print("heia")
print(accuracy_score(label, y))
```
